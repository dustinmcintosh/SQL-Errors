```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import matplotlib

font = {'family' : 'normal',
        'size'   : 24}

matplotlib.rc('font', **font)

pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 8)

%matplotlib inline

data = pd.read_csv('error_message_data_v2.csv')
```

```python
#get expert user data only (
minQueries = 1000
maxQueries = 100000000
UserQueryCount = data['executed_by_id'].value_counts()
UserQueryCount = UserQueryCount[UserQueryCount >= minQueries]
UserQueryCount = UserQueryCount[UserQueryCount < maxQueries]
expertUsers = UserQueryCount.index.values
def isexpertUser(userid):
    if userid in expertUsers:
        return True
    else:
        return False

data['isExpert'] = data['executed_by_id'].apply(isexpertUser)
expertUserData = data[data['isExpert']]
```

```python
#get intermediate user data only (
minQueries = 5
maxQueries = 1000
UserQueryCount = data['executed_by_id'].value_counts()
UserQueryCount = UserQueryCount[UserQueryCount >= minQueries]
UserQueryCount = UserQueryCount[UserQueryCount < maxQueries]
intermediateUsers = UserQueryCount.index.values
def isintermediateUser(userid):
    if userid in intermediateUsers:
        return True
    else:
        return False

data['isIntermediate'] = data['executed_by_id'].apply(isintermediateUser)
intermediateUserData = data[data['isIntermediate']]
```

```python
#get infrequent user data only (
minQueries = 1
maxQueries = 5
UserQueryCount = data['executed_by_id'].value_counts()
UserQueryCount = UserQueryCount[UserQueryCount >= minQueries]
UserQueryCount = UserQueryCount[UserQueryCount < maxQueries]
infrequentUsers = UserQueryCount.index.values
def isinfrequentUser(userid):
    if userid in infrequentUsers:
        return True
    else:
        return False

data['isInfrequent'] = data['executed_by_id'].apply(isinfrequentUser)
infrequentUserData = data[data['isInfrequent']]
```

```python
#here I put the times into python's datetime

import datetime as dt

def get_time(Modedt):
    try:
        year = Modedt[:4]
        month = Modedt[5:7]
        day  = Modedt[8:10]
        hour  = Modedt[11:13]
        minute  = Modedt[14:16]
        sec  = Modedt[17:19]
        micsec  = Modedt[20:23]
        return dt.datetime(int(year), int(month), int(day), int(hour), int(minute), int(sec), int(micsec)*1000)
    except TypeError:
        #print Modedt
        return float('nan')
    
print get_time(data['created_at'].iloc[0])

#sometimes Completed_at = nan 
intermediateUserData['createdT'] = intermediateUserData['created_at'].apply(get_time)
intermediateUserData['completedT'] = intermediateUserData['completed_at'].apply(get_time)
intermediateUserData['deltaTime'] = (intermediateUserData['completedT']-intermediateUserData['createdT'])
```

```python
font = {'family' : 'normal',
        'size'   : 24}

matplotlib.rc('font', **font)

def getYearMonth(timestamp):
    return dt.date(timestamp.year, timestamp.month, 15)

intermediateUserData['YearMonth'] = intermediateUserData['createdT'].apply(getYearMonth)
intermediateUserData['YearMonth'].value_counts()

timeofQuit = intermediateUserData['YearMonth'].value_counts()
timeofQuit = timeofQuit.sort_index()

#Number of queries of people that 
timeofQuit.plot(kind="bar", color = 'r', figsize = (8, 5))

plt.xlabel('Time')
plt.ylabel('Queries by intermediate users')
```
