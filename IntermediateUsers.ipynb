```python
#get expert user data only (
minQueries = 1000
maxQueries = 100000000
UserQueryCount = data['executed_by_id'].value_counts()
UserQueryCount = UserQueryCount[UserQueryCount >= minQueries]
UserQueryCount = UserQueryCount[UserQueryCount < maxQueries]
expertUsers = UserQueryCount.index.values
def isexpertUser(userid):
    if userid in expertUsers:
        if userid in [1513, 5613]:
            return False
        else:
            return True
    else:
        return False

data['isExpert'] = data['executed_by_id'].apply(isexpertUser)
expertUserData = data[data['isExpert']]
expertUserData = expertUserData.dropna(subset = ['raw_source'])
```

```python
#get intermediate user data only (
minQueries = 10
maxQueries = 1000
UserQueryCount = data['executed_by_id'].value_counts()
UserQueryCount = UserQueryCount[UserQueryCount >= minQueries]
UserQueryCount = UserQueryCount[UserQueryCount < maxQueries]
intermediateUsers = UserQueryCount.index.values
def isintermediateUser(userid):
    if userid in intermediateUsers:
        return True
    else:
        return False

data['isIntermediate'] = data['executed_by_id'].apply(isintermediateUser)
intermediateUserData = data[data['isIntermediate']]
intermediateUserData = intermediateUserData.dropna(subset = ['raw_source'])
```

```python
#get infrequent user data only (
minQueries = 1
maxQueries = 10
UserQueryCount = data['executed_by_id'].value_counts()
UserQueryCount = UserQueryCount[UserQueryCount >= minQueries]
UserQueryCount = UserQueryCount[UserQueryCount < maxQueries]
infrequentUsers = UserQueryCount.index.values
def isinfrequentUser(userid):
    if userid in infrequentUsers:
        return True
    else:
        return False

data['isInfrequent'] = data['executed_by_id'].apply(isinfrequentUser)
infrequentUserData = data[data['isInfrequent']]
infrequentUserData = infrequentUserData.dropna(subset = ['raw_source'])
```
