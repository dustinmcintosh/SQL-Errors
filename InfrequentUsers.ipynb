```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import matplotlib

font = {'family' : 'normal',
        'size'   : 24}

matplotlib.rc('font', **font)

pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 8)

%matplotlib inline

data = pd.read_csv('error_message_data_v2.csv')
```

```python
#get infrequent user data only (
maxQueries = 5
UserQueryCount = data['executed_by_id'].value_counts()
UserQueryCount = UserQueryCount[UserQueryCount <= maxQueries]
infrequentUsers = UserQueryCount.index.values
def isinfrequentUser(userid):
    if userid in infrequentUsers:
        return True
    else:
        return False

data['isInfrequent'] = data['executed_by_id'].apply(isinfrequentUser)
infrequentUserData = data[data['isInfrequent']]
```

```python
#frac of infrequent users that make errors
print float(len(infrequentUserData['error_message'].dropna()))/len(infrequentUserData)
```

```python
#here I put the times into python's datetime

import datetime as dt

def get_time(Modedt):
    try:
        year = Modedt[:4]
        month = Modedt[5:7]
        day  = Modedt[8:10]
        hour  = Modedt[11:13]
        minute  = Modedt[14:16]
        sec  = Modedt[17:19]
        micsec  = Modedt[20:23]
        return dt.datetime(int(year), int(month), int(day), int(hour), int(minute), int(sec), int(micsec)*1000)
    except TypeError:
        #print Modedt
        return float('nan')
    
print get_time(data['created_at'].iloc[0])

#sometimes Completed_at = nan 
infrequentUserData['createdT'] = infrequentUserData['created_at'].apply(get_time)
infrequentUserData['completedT'] = infrequentUserData['completed_at'].apply(get_time)
infrequentUserData['deltaTime'] = (infrequentUserData['completedT']-infrequentUserData['createdT'])
```

```python
font = {'family' : 'normal',
        'size'   : 24}

matplotlib.rc('font', **font)

def getYearMonth(timestamp):
    return dt.date(timestamp.year, timestamp.month, 15)

infrequentUserData['YearMonth'] = infrequentUserData['createdT'].apply(getYearMonth)
infrequentUserData['YearMonth'].value_counts()

timeofQuit = infrequentUserData['YearMonth'].value_counts()
timeofQuit = timeofQuit.sort_index()

#Number of queries of people that 
timeofQuit.plot(kind="bar", color = 'r', figsize = (8, 5))

plt.xlabel('Time')
plt.ylabel('Queries by users with '+str(maxQueries)+' or less queries')

```
