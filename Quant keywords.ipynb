```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import matplotlib

font = {'family' : 'normal',
        'size'   : 24}

matplotlib.rc('font', **font)

pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 8)

%matplotlib inline

data = pd.read_csv('error_message_data_v2.csv')
```

```python
sqlKeywords = pd.read_csv('keywords.txt', names = '1').dropna().values.tolist()
print sqlKeywords[:5]
print len(sqlKeywords)
```

```python
#df = random subset of 1000 of the errors
import random
rows = random.sample(data.index, 100000)
df = data.ix[rows]
df = df.dropna()
```

```python
def StripSQLQuery(query):
    thisQuery = query.split('\n')
    result = []
    for line in thisQuery:
        result.extend(line.split(" "))
    return result
    
StripSQLQuery(data['raw_source'].iloc[0])
```

```python
data['raw_source'].iloc[10]
```

```python
#Count each keyword in each query
for keyword in sqlKeywords:
    
    keyToSearch = keyword[0]
    #print keyToSearch
    def getCount(SQLquery):
        return SQLquery.lower().count(keyToSearch.lower())
    
    df[keyToSearch + ' freq'] = df['raw_source'].apply(getCount)
```

```python
df['SELECT' + ' freq']
```

```python
#This is not correct as "in" is a substring of "having". for example

fig = plt.figure(figsize = (8, 5))

font = {'family' : 'normal',
        'size'   : 18}
matplotlib.rc('font', **font)

frequencies = df.iloc[:, 12:].sum()
frequencies = frequencies[frequencies>0]/float(sum(frequencies))
frequencies.sort(ascending=False)

frequencies.plot(kind="bar", figsize=(25, 8))
```

```python
#Count each keyword in each query
import re

for keyword in sqlKeywords:
    
    keyToSearch = keyword[0].lower()
    #print keyToSearch
    def getCount(SQLquery):
        queryParse = re.findall(r"[\w']+|[.,!?;:)(]", SQLquery.lower())
        return queryParse.count(keyToSearch)
    
    df[keyToSearch] = df['raw_source'].apply(getCount)
```
